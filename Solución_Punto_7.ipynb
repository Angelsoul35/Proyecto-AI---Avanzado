```python
# Análisis de sentimientos

from transformers import pipeline

# Crear el pipeline para análisis de sentimiento
sentiment_analyzer = pipeline('sentiment-analysis')

# Ejemplo de texto para análisis de sentimiento
text = "I love using Hugging Face Transformers. They make NLP so easy!"

# Realizar el análisis de sentimiento
result = sentiment_analyzer(text)
print("Sentiment Analysis Result:", result)
```

```python
# Resumen de textos

# Crear el pipeline para resumen de texto
summarizer = pipeline('summarization')

# Ejemplo de texto para resumen
text = """La ciencia de datos es el estudio de datos con el fin de extraer información significativa para empresas.
Es un enfoque multidisciplinario que combina principios y prácticas del campo de las matemáticas, la estadística,
la inteligencia artificial y la ingeniería de computación para analizar grandes cantidades de datos. Este análisis
permite que los científicos de datos planteen y respondan a preguntas como “qué pasó”, “por qué pasó”, “qué pasará”
y “qué se puede hacer con los resultados”."""

# Realizar el resumen de texto
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
print("Text Summary:", summary)
```

```python
# Clasificación de documentos

# Crear el pipeline para clasificación de documentos
classifier = pipeline('zero-shot-classification')

# Ejemplo de texto para clasificación
text = "Hugging Face provides state-of-the-art machine learning models for various NLP tasks."

# Definir las etiquetas de clasificación
candidate_labels = ['technology', 'sports', 'politics', 'health']

# Realizar la clasificación de documentos
classification = classifier(text, candidate_labels)
print("Document Classification Result:", classification)
```

```python
# Traducción automática

# Crear el pipeline para traducción automática
translator = pipeline('translation_en_to_fr')

# Ejemplo de texto para traducción
text = "Hugging Face is a leader in the field of NLP."

# Realizar la traducción automática
translation = translator(text)
print("Translation Result:", translation)
```
